# EDA

 - EDA stands for Exploratory Data Analysis. It's an approach to analyzing datasets to summarize their main characteristics, often using
    statistical graphs and other data visualisation techniques.

 - The primary goal of the EDA is to understand what the data can tell us beyond the formal modelling or hypothesis testing tasks. It 
    involves looking at the data from different angles summarising it's main features and making sense of the data.

# Bivariate Analysis

 - Bivariate Analysis is a statistical method that involves the analysis of two variables simultaniously to determine the relatioship between them.
 - It focuses on understanding how changes in one variable affect changes in another variable

Graph Preferences
 - Scatterplot
 - lineplot
 - bubbleplots
 - heatmap
 - barplot
 - pie

Steps:
 - Select two columns
 - Understanding relationship between the columns
   - **Numerical-Numerical**: Check coorelations. You can plot graphs like lineplot, heatmaps
   - **Ctegorical-Numerical**: Create visualisation that compare the distribution of the numerical data across different categories of the categorical data
   - **Categorical-Categorical**: You can create a cross-tabulation of contigency tables that show the distribution of values in one categorical column grouped by the values in the other categorical column

## MultiVariant Analysis

- Multivariant Analysis is a statistical method used to analyze and understand relationship, patterns, and interactions among multiple variables withina data set simultaniously 
- It involves examining the collective influence of several variables on one another and on the outcomes of interest
